```python
import re
```

```python
m = re.findall('abc','aaaaabcaaaabc')
print(m)
```

```python
m = re.findall('\d','a1b2c3d')
print(m)
```

```python
m = re.findall('\d\d\d\d','aa123b1234df12344')
print(m)
```

```python
m = re.findall(r'<div>(.*)<div>','<div>hello<div>')
print(m)
```

```python
m = re.findall(r'<div>(.*)<div>','<div>hello<div><div>hello<div><div>hello<div>')
print(m)#贪婪
```

```python
m = re.findall(r'<div>(.*?)<div>','<div>hello<div><div>hello<div><div>hello<div>')
print(m)#非贪婪
```

### 解析一篇博客的源码

```python
import urllib.request
```

```python
res = urllib.request.urlopen('https://www.runoob.com/regexp/regexp-syntax.html')
```

```python
ret = res.read()
```

```python
ret.decode("utf8")
```

```python
response = re.findall('<title>(.*?)</title>',ret.decode('utf8'))
```

```python
print(response)
```

```python
#匹配非换行符
m = re.findall(".","aa\nbbcc")
print (m)
```

```python
#匹配转义字符
m = re.findall("\.","a.c")
print(m)
```

```python
#匹配字符集
m = re.findall("a[bcd]e","abeaceade")
print(m)
```

```python
#匹配数字
m = re.findall("\d","a1b")
print(m)
```

```python
#匹配非数字
m = re.findall("\D","a1b")
print(m)
```

```python
#匹配空白字符
m = re.findall("\s","a 1\tb")
print(m)
```

```python
#匹配非空白字符
m = re.findall("\S","a 1\tb")
print(m)
```

```python
#匹配字母数字
m = re.findall("\w","ab#12--")
print(m)
```

```python
#匹配非字母数字
m = re.findall("\W","ab#12--")
print(m)
```

```python
#匹配开头
m = re.findall("^abc","abcabc")
print(m)
```

```python
#匹配结尾
m = re.findall("abc$","abcabc")
print(m)
```

```python
#不区分大小写
m = re.findall("abc","abcdABc",re.I)
print(m)
```

```python
#匹配换行
m = re.findall("<div>(.*?)<div>","<div>hello\nworld<div>")
print(m)
m = re.findall("<div>(.*?)<div>","<div>hello\nworld<div>",re.S)
print(m)
```

```python
#匹配换行\不区分大小写
m = re.findall("<div>(.*?)<div>","<div>hello\nworld<Div>",re.S)
print(m)
m = re.findall("<div>(.*?)<div>","<div>hello\nworld<Div>",re.S|re.I)
print(m)
```

```python
#匹配一个或零个
m = re.findall('ab?','abbbaba')
print(m)
```

```python
#匹配一个或多个
m = re.findall('ab+','abbbaba')
print(m)
```

```python
#匹配0个或多个
m = re.findall('ab*','abbbaba')
print(m)
```

```python
#匹配以org结尾的邮箱
m = re.findall('\w+@\w+\.org','123456@qq.com;123456@qq.org')
print(m)
```
